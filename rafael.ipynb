```python

#         "3.1. Vamos a realizar un bucle for donde iteraremos por los diferentes valores de la columna ``'Gls'`` para crear una nueva columna que se llamara ``'type_goal_player'``. Dejo a tu elección los valores en base dónde cortar para crear las diferentes categorías. Hay que crear tres valor Poco goleador, goleador, Muy goleador\n"


df301=df

df['type_goal_player']=''


for index, row in df.iterrows():
    if row['Gls'] == 0:  
        df.at[index, 'type_goal_player'] = 'Poco Goleador' 
    elif 1 <= row['Gls'] <= 4: 
        df.at[index, 'type_goal_player'] = 'Goleador' 
    else: 
        df.at[index, 'type_goal_player'] = 'Muy Goleador'


display(df301)
```

```python


df3011 = df301.loc[:, ['Player','Gls', 'type_goal_player']]
df3011= df3011.sort_values(by='Gls',ascending=False)
display(df3011)
```

```python

#         "3.2. Filtramos los datos para los jugadores que tienen el valor de ``'Muy goleador'`` en nuestra nueva columna."

df302 = df3011.loc[df3011['type_goal_player'] == 'Muy Goleador', ['Player', 'Gls', 'type_goal_player']]



df302=df302.sort_values(by='Gls',ascending=False)
display(df302)
```

```python

#         "3.3. Quiero saber la cantidad de jugadores de cada tipo de goleador. Cuantas veces sale cada valor en la columna ``'type_goal_player'``."

tabela303= df3011.pivot_table(index='type_goal_player', columns='type_goal_player',aggfunc='size')


display(tabela303)
```
